```python
# find_many_solutions.py
import random
from sympy import isprime
```

```python
def generate_random_start(exp_min=70, exp_max=75):
    """ç”Ÿæˆ 10^exp_min åˆ° 10^exp_max ä¹‹é—´çš„éšæœºå¤§æ•´æ•°"""
    exp = random.randint(exp_min, exp_max)
    digits = [str(random.randint(1, 9))]  # é¦–ä½éé›¶
    digits += [str(random.randint(0, 9)) for _ in range(exp - 1)]
    return int(''.join(digits))

def count_primes_in_window(start, length=2004):
    """å¿«é€Ÿè®¡æ•°çª—å£å†…è´¨æ•°"""
    count = 0
    for i in range(length):
        if isprime(start + i):
            count += 1
            if count > 12:  # æ—©æœŸé€€å‡º
                return count
    return count
```

```python
def main():
    solutions = []
    attempts = 0
    print("ğŸ” åœ¨ 10^70 ~ 10^75 èŒƒå›´å†…æœç´¢è§£...")
    
    while len(solutions) < 10:  # æ‰¾ 10 ä¸ªè§£
        attempts += 1
        N = generate_random_start()
        count = count_primes_in_window(N)
        
        if count == 12:
            solutions.append(N)
            print(f"âœ… è§£ #{len(solutions)}: {N}")
            with open("solutions.txt", "a") as f:
                f.write(f"{N}\n")
        
        if attempts % 100 == 0:
            print(f"  å·²å°è¯• {attempts} æ¬¡, æ‰¾åˆ° {len(solutions)} ä¸ªè§£")

if __name__ == "__main__":
    main()
```
