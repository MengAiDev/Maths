```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

import sys
from sympy import isprime, factorint, nextprime, prevprime
from math import isqrt

# ========== è¾“å…¥ ==========
N = 283652129125808400513278476301455085008845288816557395539337194639631785
LENGTH = 2004
H = [166, 278, 604, 724, 1096, 1256, 1306, 1492, 1514, 1564, 1574, 1712]
S = [j for j in range(LENGTH) if j not in H]

print("=" * 80)
print(f"éªŒè¯èµ·å§‹æ•° N (å…± {len(str(N))} ä½):")
print(f"N = {N}")
print("=" * 80)

# ========== æ­¥éª¤1: éªŒè¯ H ä¸­çš„ä½ç½®æ˜¯å¦å…¨ä¸ºè´¨æ•° ==========
print("\nã€æ­¥éª¤1ã€‘æ£€æŸ¥ H ä¸­çš„ 12 ä¸ªä½ç½®æ˜¯å¦ä¸ºè´¨æ•°...")
prime_positions = []
for h in H:
    n_val = N + h
    print(f"  æ£€æŸ¥ N + {h:2d} = ...{str(n_val)[-20:]} ", end="", flush=True)
    if isprime(n_val):
        prime_positions.append(h)
        print("âœ… PRIME")
    else:
        print("âŒ NOT PRIME")
        sys.exit(1)

if len(prime_positions) != 12:
    print("âŒ é”™è¯¯ï¼šH ä¸­è´¨æ•°æ•°é‡ â‰  12")
    sys.exit(1)

print("âœ… æ­¥éª¤1é€šè¿‡ï¼šH ä¸­ 12 ä¸ªæ•°å‡ä¸ºè´¨æ•°ã€‚")

# ========== æ­¥éª¤2: éªŒè¯ S ä¸­çš„ä½ç½®æ˜¯å¦å…¨ä¸ºåˆæ•° ==========
print("\nã€æ­¥éª¤2ã€‘æ£€æŸ¥ S ä¸­çš„ 1992 ä¸ªä½ç½®æ˜¯å¦ä¸ºåˆæ•°...")
composite_count = 0
failed_positions = []

# å°è¯•å¿«é€Ÿæ‰¾å°å› å­ï¼ˆåŠ é€Ÿåˆæ•°ç¡®è®¤ï¼‰
def has_small_factor(n, limit=1000000):
    """è¿”å›ä¸€ä¸ªå°å› å­ï¼ˆè‹¥å­˜åœ¨ï¼‰ï¼Œå¦åˆ™è¿”å› None"""
    if n % 2 == 0:
        return 2
    # è¯•é™¤å¥‡æ•°åˆ° limit
    f = 3
    while f * f <= n and f <= limit:
        if n % f == 0:
            return f
        f += 2
    return None

for idx, j in enumerate(S):
    if idx % 200 == 0:
        print(f"  è¿›åº¦: {idx}/1992", file=sys.stderr, flush=True)
    
    n_val = N + j
    # å¿«é€Ÿæ£€æŸ¥æ˜¯å¦ä¸ºè´¨æ•°ï¼ˆåº”ä¸º Falseï¼‰
    if isprime(n_val):
        failed_positions.append(j)
        print(f"  âŒ æ„å¤–å‘ç°è´¨æ•°: N + {j} æ˜¯è´¨æ•°!", file=sys.stderr)
        continue

    # å°è¯•æ‰¾å°å› å­ä»¥åŠ é€Ÿç¡®è®¤
    small_factor = has_small_factor(n_val, limit=1000000)
    if small_factor:
        composite_count += 1
        # å¯é€‰ï¼šæ‰“å°å› å­ï¼ˆæ³¨é‡Šæ‰ä»¥å‡å°‘è¾“å‡ºï¼‰
        # print(f"  N + {j} = ... åˆæ•° (å› å­: {small_factor})")
    else:
        # æ— å°å› å­ï¼Œä½† isprime å·²è¿”å› False â†’ ä»ä¸ºåˆæ•°ï¼ˆå¯èƒ½å¤§å› å­ï¼‰
        composite_count += 1

if failed_positions:
    print(f"\nâŒ å¤±è´¥ï¼åœ¨éHä½ç½®å‘ç° {len(failed_positions)} ä¸ªæ„å¤–è´¨æ•°:")
    for j in failed_positions[:10]:
        print(f"  N + {j}")
    if len(failed_positions) > 10:
        print(f"  ... è¿˜æœ‰ {len(failed_positions)-10} ä¸ª")
    sys.exit(1)

if composite_count != len(S):
    print(f"âŒ åˆæ•°è®¡æ•°é”™è¯¯: æ‰¾åˆ° {composite_count}, åº”ä¸º {len(S)}")
    sys.exit(1)

print("âœ… æ­¥éª¤2é€šè¿‡ï¼šS ä¸­ 1992 ä¸ªæ•°å‡ä¸ºåˆæ•°ã€‚")

# ========== æ­¥éª¤3: æœ€ç»ˆç»“è®º ==========
print("\n" + "="*80)
print("ğŸ‰ éªŒè¯æˆåŠŸï¼")
print(f"åœ¨ [{N}, {N + LENGTH - 1}] çš„è¿ç»­ {LENGTH} ä¸ªæ•´æ•°ä¸­ï¼š")
print(f"  - è´¨æ•°æ•°é‡: {len(prime_positions)}")
print(f"  - è´¨æ•°ä½ç½®: {sorted(prime_positions)}")
print(f"  - ä¸ç›®æ ‡ H = {H} å®Œå…¨ä¸€è‡´")
print("="*80)

# å¯é€‰ï¼šè¾“å‡ºè´¨æ•°åˆ—è¡¨ï¼ˆç”¨äºæäº¤è®°å½•ï¼‰
print("\nã€é™„ã€‘12 ä¸ªè´¨æ•°ï¼ˆå®Œæ•´å€¼ï¼‰:")
for h in H:
    print(N + h)
```
